### Plot the heat map


```python
#plotting heatmap to correlate the features
correlate=df.corr()
sns.heatmap(correlate,xticklabels=correlate.columns,yticklabels=correlate.columns,annot=True)
```

### Plot a Scatter Plot 

```python
# We plot this to see the relationship between sales and profit
plt.scatter(x=df['Sales'],y=df['Profit'])
plt.xlabel("Overall Sales")
plt.ylabel("Profit")
plt.show()
```

```python
# We plot this to see the relationship between discount and profit
plt.scatter(x=df['Discount'],y=df['Profit'])
plt.xlabel("Overall Discount")
plt.ylabel("Profit")
plt.show()
```

### Compare the data 

```python
#compare ship mode
sns.set_style('whitegrid')
sns.countplot(x='Ship Mode',data=df,palette='BrBG')
```

```python
#compare segment
sns.set_style('whitegrid')
sns.countplot(x='Segment',data=df,palette='BrBG')
```

```python
#Compare Region
sns.set_style('whitegrid')
sns.countplot(x='Region',data=df,palette='BrBG')
```

```python
#Compare Sub_category
sns.set_style('whitegrid')
sns.countplot(x='Sub-Category',data=df,palette='BrBG')
```

```python
# Compare segment and region 
sns.set_style('whitegrid')
sns.countplot(x='Ship Mode',hue='Region',data=df,palette='cubehelix')
```

```python
#compare segmment and category
sns.set_style('whitegrid')
sns.countplot(x='Segment',hue='Category',data=df,palette='cubehelix')
```

```python
#compare region and category
sns.countplot(x='Region',hue='Category',data=df,palette='cubehelix')
```
